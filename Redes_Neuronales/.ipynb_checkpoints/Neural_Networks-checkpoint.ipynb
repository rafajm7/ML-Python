# Ejercicio 1

```python
import numpy as np
from __future__ import division
import math

W = {1: np.array([[-2,1], [1,-1], [3,-1]]), 2: np.array([2,3,1])}
s = np.array([2,3,1])
b = {1: np.array([0,-1,1]), 2: 0}
a = {1: np.array([1,1])}
z = {}
```

### Apartado a

Para este apartado comenzamos definiendo la función sigmoide, la cual es la función de activación en todas las neuronas salvo en las de salida.

```python
def sigmoide(x):
    return 1 / (1 + math.exp(-x))
```

Calculamos `z` y `a` para todas las neuronas. En el caso de que `z` sea un solo número, quiere decir que estamos en la capa de salida, y usamos la función identidad.

```python
for i in range(2,4):
    z[i] = np.matmul(W[i-1], a[i-1]) + b[i-1]
    if isinstance(z[i],float):
        a[i] = z[i]
    else:
        a[i] = np.array(map(sigmoide, z[i]))
```

A continuación mostramos los valores obtenidos para `z` y `a`:

```python
print z
print a
```

### Apartado b

Comenzamos el apartado definiendo la derivada de la sigmoide, para la que se cumple que $f'(z) = f(z)*(1-f(z))$:

```python
def deriv_sigmoide(x):
    return sigmoide(x)*(1-sigmoide(x))
```

Para calcular $\delta$, seguimos las fórmulas vistas en clase:

```python
delta = {}
delta[3] = -(1 - a[3])*1
deriv = np.array(map(deriv_sigmoide, z[2]))
print deriv
# en este caso hemos de hacer un producto escalar, y por eso utilizamos np.multiply en lugar de np.matmul
delta[2] = np.array([np.multiply(W[2]*delta[3],deriv)])
```

Mostramos los valores obtenidos para delta:

```python
delta
```

### Apartado c

Comenzamos el apartado calculando las derivadas parciales de W y b, para posteriormente usarlas en el recálculo de W y b.

```python
alpha = 0.5
lambda_ = 1
part_W = {1: np.multiply(delta[2].transpose(),a[1]), 2: np.multiply(delta[3],a[2])}
part_b = {1: delta[2], 2: delta[3]}
```

```python
print part_W
print part_b
```

```python
for i in range(1,3):
    W[i] = W[i] - alpha*(part_W[i]+lambda_*W[i])
    b[i] = b[i] - alpha*part_b[i]
```

```python
print W
print b
```

# Ejercicio 2

Importamos las librerías necesarias para la realización de los ejercicios 2 y 3.

```python
from sklearn import preprocessing, cross_validation, neural_network
from sklearn.grid_search import GridSearchCV
from sklearn.preprocessing import Imputer
from sklearn.externals import joblib
import sklearn.metrics as metrics

import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from operator import itemgetter
```

Cargamos el fichero y definimos los conjuntos de train y test.

```python
np.random.seed(1)
datasetName = '../datos/bloodTransfusion.data'
datasetDelimiter = ','

# load the CSV file as a numpy matrix
dataset = np.loadtxt(datasetName, delimiter=datasetDelimiter)
xRaw = dataset[:,0:dataset.shape[1]-1]
y = dataset[:,dataset.shape[1]-1]
y=y.astype('int')

#Divide in training and test, shuffling the examples and keeping the proportion of examples of each class
xTrain_tmp, xTest_tmp, yTrain, yTest = cross_validation.train_test_split(xRaw, y, test_size=0.2)

#Standardize data
scaler = preprocessing.StandardScaler().fit(xTrain_tmp)
xTrain=scaler.transform(xTrain_tmp)
xTest=scaler.transform(xTest_tmp)
```

Procedemos a realizar el entrenamiento para los valores de los hiperparámetros que se muestran a continuación. Una vez los tenemos, creamos dos diccionarios que contendrán el error y la desviación estándar, respectivamente, de cada una de las combinaciones de los hiperparámetros.

```python
#Generate grid search
hidden_layer_sizes = []
capasList = [1,2,3]
sizes = list(np.arange(5,105,5))
alphas = [5e-5, 0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1]
for capas in capasList:
    for size in sizes:
        hidden_layer_sizes.append((size,)*capas)
    
hyperParams = {'hidden_layer_sizes': hidden_layer_sizes, 'alpha': alphas}

modelCV = GridSearchCV(neural_network.MLPClassifier(activation='tanh', solver='lbfgs'), 
                     hyperParams, cv=5, scoring='accuracy')
modelCV.fit(xTrain, yTrain)
print "Best hyperparameter", modelCV.best_params_
print "\n\n"

errores = { "Ncapas:"+str(capas)+"-Alpha:"+str(alpha):[] for capas in capasList for alpha in alphas}
devs = { "Ncapas:"+str(capas)+"-Alpha:"+str(alpha):[] for capas in capasList for alpha in alphas}
for hyperP, mean_score, scores in modelCV.grid_scores_:
    print("%0.3f (+/-%0.03f) for %r"
              % (mean_score, scores.std(), hyperP))
    errores["Ncapas:"+str(len(hyperP['hidden_layer_sizes']))+"-Alpha:"+str(hyperP['alpha'])].append(1-mean_score)
    devs["Ncapas:"+str(len(hyperP['hidden_layer_sizes']))+"-Alpha:"+str(hyperP['alpha'])].append(scores.std())
```

Gráficas con el error de validación cruzada frente al número de nodos. Hemos decidido mostrarlas en grupos de dos, para lo que hemos decidido definir una función llamada `pairwise` que, dada una lista, nos devuelve la misma en pares de valores, lo que nos permite recorrerla de dos en dos.

```python
from itertools import izip

def pairwise(iterable):
    "s -> (s0, s1), (s2, s3), (s4, s5), ..."
    a = iter(iterable)
    return izip(a, a)
```

```python
for capas in capasList:
    for alpha1, alpha2 in pairwise(alphas):
        fig, axs = plt.subplots(ncols=2, figsize=(15.5, 5.5))
        axs[0].set_title("Ncapas="+str(capas)+", Alpha="+str(alpha1))
        axs[1].set_title("Ncapas="+str(capas)+", Alpha="+str(alpha2))
        
        axs[0].errorbar(sizes, errores["Ncapas:"+str(capas)+"-Alpha:"+str(alpha1)], yerr=devs["Ncapas:"+str(capas)+"-Alpha:"+str(alpha1)])
        plt.xlim(sizes[0]-10, sizes[len(sizes)-1]+10)
        
        axs[1].errorbar(sizes, errores["Ncapas:"+str(capas)+"-Alpha:"+str(alpha2)], yerr=devs["Ncapas:"+str(capas)+"-Alpha:"+str(alpha2)])
        plt.xlim(sizes[0]-10, sizes[len(sizes)-1]+10)
```

A continuación mostramos el menor error de entrenamiento, la desviación estándar y el valor del hiperparámetro con el que se corresponde.

```python
##Menor error CV
best_param_cv_nodos = modelCV.best_params_["hidden_layer_sizes"][0]
best_param_cv_ncapas = len(modelCV.best_params_["hidden_layer_sizes"])
best_param_cv_alpha = modelCV.best_params_["alpha"]
pos_nodos = sizes.index(best_param_cv_nodos)
print "Mejores parámetros : Alpha = " , best_param_cv_alpha, ", Ncapas = ", best_param_cv_ncapas, ", Nnodos =  ", best_param_cv_nodos
print "Error de entrenamiento: ", errores["Ncapas:"+str(best_param_cv_ncapas)+"-Alpha:"+str(best_param_cv_alpha)][pos_nodos]
print "Std: ", devs["Ncapas:"+str(best_param_cv_ncapas)+"-Alpha:"+str(best_param_cv_alpha)][pos_nodos]
```

Obtenemos a continuación la predicción en test para cada valor de los hiperparámetros, y guardamos su error.

```python
#Test error
testErrList={ "Ncapas:"+str(capas)+"-Alpha:"+str(alpha):[] for capas in capasList for alpha in alphas}
for alpha in alphas:
    for capas in capasList:
        for size in sizes:
            hls = (size,)*capas
            model = neural_network.MLPClassifier(activation='tanh', solver='lbfgs', alpha=alpha, hidden_layer_sizes=hls)
            model.fit(xTrain, yTrain)
            yPred=model.predict(xTest)
            testErrList["Ncapas:"+str(capas)+"-Alpha:"+str(alpha)].append(1-metrics.accuracy_score(yTest,yPred))
```

A continuación mostramos las gráficas del error de test comparadas con el de entrenamiento.

```python
for capas in capasList:
    for alpha1, alpha2 in pairwise(alphas):
        fig, axs = plt.subplots(ncols=2, figsize=(15.5, 5.5))
        axs[0].set_title("Ncapas="+str(capas)+", Alpha="+str(alpha1))
        axs[1].set_title("Ncapas="+str(capas)+", Alpha="+str(alpha2))
        
        axs[0].errorbar(sizes, errores["Ncapas:"+str(capas)+"-Alpha:"+str(alpha1)], yerr=devs["Ncapas:"+str(capas)+"-Alpha:"+str(alpha1)])
        plt.xlim(sizes[0]-1, sizes[len(sizes)-1]+1)
        axs[0].errorbar(sizes, testErrList["Ncapas:"+str(capas)+"-Alpha:"+str(alpha1)]) 

        axs[1].errorbar(sizes, errores["Ncapas:"+str(capas)+"-Alpha:"+str(alpha2)], yerr=devs["Ncapas:"+str(capas)+"-Alpha:"+str(alpha2)])
        plt.xlim(sizes[0]-1, sizes[len(sizes)-1]+1)
        axs[1].errorbar(sizes, testErrList["Ncapas:"+str(capas)+"-Alpha:"+str(alpha2)]) 
```

Menor error de test y errores de test para el mejor hiperparámetro de la cross validación:

```python
best_params_test = min([(k, sizes[np.argmin(testErrList[k])], min(testErrList[k])) for k in testErrList], key=itemgetter(2))
params = best_params_test[0].split(":")
best_param_test_ncapas = params[1][0]
best_param_test_alpha = params[2]
print "Menor error de test: ", best_params_test[2]
print "Conseguido para los parámetros Ncapas =", best_param_test_ncapas, ", Nnodos =", best_params_test[1], ",Alpha =", best_param_test_alpha
print "Error de test para los parámetros Ncapas = ", best_param_cv_ncapas, ", Nnodos = ",best_param_cv_nodos,"y Alpha =",best_param_cv_alpha, "elegidos por la CV:", testErrList["Ncapas:"+str(best_param_cv_ncapas)+"-Alpha:"+str(best_param_cv_alpha)][pos_nodos]
```

# Ejercicio 3

Fijamos la semilla y dividimos en train/test como en el ejercicio anterior.

```python
np.random.seed(1)
datasetName = '../datos/EnergyEfficiency.data'
datasetDelimiter = ','

# load the CSV file as a numpy matrix
dataset_energy = np.loadtxt(datasetName, skiprows=1, delimiter=datasetDelimiter)

xRaw2 = dataset_energy[:,0:dataset_energy.shape[1]-1]
y2 = dataset_energy[:,dataset_energy.shape[1]-1]

xTrain_tmp2, xTest_tmp2, yTrain_tmp2, yTest_tmp2 = cross_validation.train_test_split(xRaw2, y2, test_size=0.2)

#Standardize data
scalerX = preprocessing.StandardScaler().fit(xTrain_tmp2)
xTrain2=scalerX.transform(xTrain_tmp2)
xTest2=scalerX.transform(xTest_tmp2)

scalerY = preprocessing.StandardScaler().fit(yTrain_tmp2)
yTrain2=scalerY.transform(yTrain_tmp2)
yTest2=scalerY.transform(yTest_tmp2)
```

El proceso de entrenamiento es el mismo que en el caso anterior, pero utilizando el algoritmo `MLPRegressor`, siendo la medida de error el error cuadrático medio, que por defecto viene en `scikit-learn` en negativo. Por ello, cuando añadimos el error a la lista de errores, le cambiamos el signo. Además, en este caso, al haber estandarizado la variable de salida, hay que desestandarizar el MSE obtenido multiplicando por la desviación típica al cuadrado, tal y como hemos visto en clase.

```python
#Generate grid search
hidden_layer_sizes2 = []
capasList2 = [1,2,3]
sizes2 = list(np.arange(5,105,5))
alphas2 = [5e-5, 0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1]
for size in sizes2:
    for capas in capasList2:
        hidden_layer_sizes2.append((size,)*capas)

    
hyperParams2 = {'hidden_layer_sizes': hidden_layer_sizes2, 'alpha': alphas2}

modelCV2 = GridSearchCV(neural_network.MLPRegressor(activation='tanh', solver='lbfgs'), 
                     hyperParams2, cv=5, scoring='neg_mean_squared_error')
modelCV2.fit(xTrain2, yTrain2)        
    
print "Best hyperparameter", modelCV2.best_params_
print "\n\n"

errores2 = { "Ncapas:"+str(capas)+"-Alpha:"+str(alpha):[] for capas in capasList2 for alpha in alphas2}
devs2 = { "Ncapas:"+str(capas)+"-Alpha:"+str(alpha):[] for capas in capasList2 for alpha in alphas2}
for hyperP, mean_score, scores in modelCV2.grid_scores_:
    mean_score_dest = -mean_score*np.var(yTrain_tmp2)
    std_dest = scores.std()*np.var(yTrain_tmp2)
    print("%0.3f (+/-%0.03f) for %r"
              % (mean_score_dest, scores.std(), hyperP))
    errores2["Ncapas:"+str(len(hyperP['hidden_layer_sizes']))+"-Alpha:"+str(hyperP['alpha'])].append(mean_score_dest)
    devs2["Ncapas:"+str(len(hyperP['hidden_layer_sizes']))+"-Alpha:"+str(hyperP['alpha'])].append(std_dest)
```

Gráficas del error de entrenamiento:

```python
for capas in capasList2:
    for alpha1, alpha2 in pairwise(alphas2):
        fig, axs = plt.subplots(ncols=2, figsize=(15.5, 5.5))
        axs[0].set_title("Ncapas="+str(capas)+", Alpha="+str(alpha1))
        axs[1].set_title("Ncapas="+str(capas)+", Alpha="+str(alpha2))
        
        axs[0].errorbar(sizes2, errores2["Ncapas:"+str(capas)+"-Alpha:"+str(alpha1)], yerr=devs2["Ncapas:"+str(capas)+"-Alpha:"+str(alpha1)])
        plt.xlim(sizes2[0]-10, sizes2[len(sizes2)-1]+10)
        
        axs[1].errorbar(sizes2, errores2["Ncapas:"+str(capas)+"-Alpha:"+str(alpha2)], yerr=devs2["Ncapas:"+str(capas)+"-Alpha:"+str(alpha2)])
        plt.xlim(sizes2[0]-10, sizes2[len(sizes2)-1]+10)
```

Menor error de entrenamiento, su hiperparámetro y su desviación típica:

```python
##Menor error CV
best_param_cv_nodos2 = modelCV2.best_params_["hidden_layer_sizes"][0]
best_param_cv_ncapas2 = len(modelCV2.best_params_["hidden_layer_sizes"])
best_param_cv_alpha2 = modelCV2.best_params_["alpha"]
pos_nodos2 = sizes2.index(best_param_cv_nodos2)
print "Mejores parámetros : Alpha = " , best_param_cv_alpha2, ", Ncapas = ", best_param_cv_ncapas2, ", Nnodos =  ", best_param_cv_nodos2
print "Error de entrenamiento: ", errores2["Ncapas:"+str(best_param_cv_ncapas2)+"-Alpha:"+str(best_param_cv_alpha2)][pos_nodos2]
print "Std: ", devs2["Ncapas:"+str(best_param_cv_ncapas2)+"-Alpha:"+str(best_param_cv_alpha2)][pos_nodos2]
```

Obtenemos el error de test para cada hiperparámetro.

```python
#Test error
testErrList2={ "Ncapas:"+str(capas)+"-Alpha:"+str(alpha):[] for capas in capasList2 for alpha in alphas2}
for alpha in alphas2:
    for capas in capasList2:
        for size in sizes2:
            hls = (size,)*capas
            model = neural_network.MLPRegressor(activation='tanh', solver='lbfgs', alpha=alpha, hidden_layer_sizes=hls)
            model.fit(xTrain2, yTrain2)
            yPred=model.predict(xTest2)
            MSE = metrics.mean_squared_error(yTest2,yPred)*np.var(yTrain_tmp2)
            testErrList2["Ncapas:"+str(capas)+"-Alpha:"+str(alpha)].append(MSE)
```

Gráficas con el error de test y el error de entrenamiento comparado con el de test, con respecto al valor del número de nodos.

```python
for capas in capasList2:
    for alpha1, alpha2 in pairwise(alphas2):
        fig, axs = plt.subplots(ncols=2, figsize=(15.5, 5.5))
        axs[0].set_title("Ncapas="+str(capas)+", Alpha="+str(alpha1))
        axs[1].set_title("Ncapas="+str(capas)+", Alpha="+str(alpha2))
        
        axs[0].errorbar(sizes2, errores2["Ncapas:"+str(capas)+"-Alpha:"+str(alpha1)], yerr=devs2["Ncapas:"+str(capas)+"-Alpha:"+str(alpha1)])
        plt.xlim(sizes2[0]-1, sizes2[len(sizes2)-1]+1)
        axs[0].errorbar(sizes2, testErrList2["Ncapas:"+str(capas)+"-Alpha:"+str(alpha1)]) 

        axs[1].errorbar(sizes2, errores2["Ncapas:"+str(capas)+"-Alpha:"+str(alpha2)], yerr=devs2["Ncapas:"+str(capas)+"-Alpha:"+str(alpha2)])
        plt.xlim(sizes2[0]-1, sizes2[len(sizes2)-1]+1)
        axs[1].errorbar(sizes2, testErrList2["Ncapas:"+str(capas)+"-Alpha:"+str(alpha2)]) 
```

Mostramos el menor error de test, y el error de test para el mejor hiperparámetro de la cross validación.

```python
best_params_test2 = min([(k, sizes2[np.argmin(testErrList2[k])], min(testErrList2[k])) for k in testErrList2], key=itemgetter(2))
params2 = best_params_test2[0].split(":")
best_param_test_ncapas2 = params2[1][0]
best_param_test_alpha2 = params2[2]
print "Menor error de test: ", best_params_test2[2]
print "Conseguido para los parámetros Ncapas =", best_param_test_ncapas2, ", Nnodos =", best_params_test2[1], ", Alpha =", best_param_test_alpha2
print "Error de test para los parámetros Ncapas =", best_param_cv_ncapas2, ", Nnodos =", best_param_cv_nodos2, "y Alpha =" , best_param_cv_alpha2,"elegidos por la CV:", testErrList2["Ncapas:"+str(best_param_cv_ncapas2)+"-Alpha:"+str(best_param_cv_alpha2)][pos_nodos2]
```
